## UNIVARIATE ANALYSIS

* Univariate Analysis means analyzing one variable at a time to understand how the data is distributed. It helps us see values like minimum, maximum, average, and detect outliers.
* There are two types of univariate analysis
  * Numerical Univariate Analysis-Used for numerical data (mean, median, range, histogram).
  * Categorical Univariate Analysis-Used for numerical data (mean, median, range, histogram).

### HISTPLOT

```python
plt.figure(figsize=(7,5))
sns.histplot(df['CGPA'], kde=True)
plt.title("Distribution of CGPA")
plt.xlabel("CGPA")
plt.show()
```

### COUNTPLOT

```python
plt.figure(figsize=(10,4))
sns.countplot(x='Result Description', data=df)
plt.title("Result Description Distribution")
plt.show()
```

## BIVARIATE ANALYSIS (Feature VS Target)

* Bivariate Analysis means analyzing two variables at the same time to understand the relationship between them. It helps us see how one variable changes with respect to another.
* Types of Bivariate Analysis
  * Numerical vs Numerical: Used to understand the relationship or correlation between two numerical variables.

  * Categorical vs Numerical: Used to compare numerical values across different categories.

  *  Categorical vs Categorical: Used to analyze the relationship and frequency between two categorical variables.

### SCATTERPLOT

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='CGPA', y='SGPA', hue='Semester', data=df)
plt.title("CGPA vs SGPA By Semester")
plt.show()
```

### BOXPLOT

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Semester', y='CGPA', hue='Semester', data=df)
plt.title("Semester vs CGPA by Semester")
plt.show()
```

## CORRELATION

### HEATMAP

```python
plt.figure(figsize=(10,4))
sns.heatmap(df.drop(columns=['Roll No']).corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## OUTLIERS DETECTION

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df['SGPA'])
plt.title("Outliers in Weight")
plt.show()
```
